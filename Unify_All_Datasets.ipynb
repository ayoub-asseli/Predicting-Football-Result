```python
import pandas as pd
```

```python
leagues = ['PL', 'SerieA', 'Bundesliga', 'Ligue1', 'Liga']
seasons = ['05-06', '06-07','07-08', '08-09', '09-10', '10-11', '11-12', '12-13', '13-14', 
           '14-15', '15-16', '16-17', '17-18', '18-19', '19-20', '20-21']
#We have data from every seasons from 2005
```

```python
pl = pd.read_csv('Data/saison-21-22-PL.csv')
ligue1 = pd.read_csv('Data/saison-05-06-Ligue1.csv')
liga = pd.read_csv('Data/saison-05-06-Liga.csv')
bunde = pd.read_csv('Data/saison-05-06-Bundesliga.csv')
serie_a = pd.read_csv('Data/saison-05-06-SerieA.csv')

com_col = []
for elem in pl.columns:
    if elem in ligue1.columns and elem in liga.columns and elem in bunde.columns and elem in serie_a:
        com_col.append(elem)
```

```python
print(com_col)
```

```python
print(len(com_col))
```

```python
i = 0
for season in seasons:
    pl = pd.read_csv('Data/saison-{}-PL.csv'.format(season))[com_col]
    ligue1 = pd.read_csv('Data/saison-{}-Ligue1.csv'.format(season))[com_col]
    liga = pd.read_csv('Data/saison-{}-Liga.csv'.format(season))[com_col]
    bunde = pd.read_csv('Data/saison-{}-Bundesliga.csv'.format(season))[com_col]
    serie_a = pd.read_csv('Data/saison-{}-SerieA.csv'.format(season))[com_col]
    pl['League'] = len(pl)*['Premier League']
    ligue1['League'] = len(ligue1)*['Ligue 1']
    liga['League'] = len(liga)*['Liga']
    bunde['League'] = len(bunde)*['Bundesliga']
    serie_a['League'] = len(serie_a)*['Serie A']
    frames = [pl, liga, serie_a, bunde, ligue1]
    if i == 0:
        df = pd.concat(frames)
        i += 1
        continue
    else:
        frames = [df, pl, liga, serie_a, bunde, ligue1]
        df = pd.concat(frames)
df
```

```python
df = df.drop(columns=['Div'])
df = df.reset_index()
del df['index']
df
```

```python
df.to_csv('global_dataset.csv')
```
